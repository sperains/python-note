## 文件读写

### 读文件
    - 通过python内置的open()方法来打开一个文件对象
    - 文件打开成功，通过read()方法一次性读取文件的内容
    - 最后通过close()方法关闭文件
    - open()方法后面加上'r' 表示只读
    
### 写文件
    - open()方面后面加上'w'表示写，但，此时文件原有的内容将被覆盖
    - 'a' 表示追加内容


```python
f = open('C:\\Users\\Administrator\\Desktop\\新建文本文档.txt','r')

print(f)

d = f.readlines()
for line in d:
    print(line.strip())
```

## 操作文件和目录


```python
#获取当前目录的绝对路径
current = os.path.abspath('.')
print(current)

# 创建一个新目录需要把新目录的完成了路径表示出来
# 把两个路径合成一个时，不要直接拼写字符串，
# 通过os.path.join()函数可以正确处理不同操作系统的路径分隔符。
nd = os.path.join(current,'testdir')
# os.mkdir(nd)

# 删除文件
# os.rmdir(nd)

# 拆分路径。返回一个tuple，后一部分总是最后级别的目录或文件名。
d = os.path.split(nd)
print(d)

# 直接获取文件扩展名
ext = os.path.splitext('/path/to/file.txt')
print(ext)
print(ext[-1])

# 获取当前目录下的所有文件夹
dirs = [ x for x in os.listdir('.') if os.path.isdir(x)]
print(dirs)
```
